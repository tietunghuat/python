```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
from PIL import Image
```

read in file

```python
text = open("cloud.txt").read()
```

generate word cloud

```python
wordcloud = WordCloud().generate(text)

```

show the cloud

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(width=600, 
                      height=600,
                      max_words=10,
                      prefer_horizontal=0.5,
                      background_color="rgba(0,0,255,0)", 
                      mode="RGBA").generate(text)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file("wordcloud.png")
```

```python
dove_mask = np.array(Image.open("xmas_tree.png"))
#dove_mask[230:250, 240:250]
print(dove_mask)
plt.imshow(dove_mask)
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(
                      mask=dove_mask,
                      contour_width=3, 
                      max_words=20,
                      repeat=True,
                      prefer_horizontal=0.5,
                      min_font_size=3,
                      background_color="rgba(0,255,0,255)",  
                      contour_color='darkgreen')

# Generate a wordcloud
wordcloud.generate(text)

# store to file
wordcloud.to_file("dove_wordcloud.png")

# show

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
